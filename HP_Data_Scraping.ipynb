```python
data = pd.read_csv('final_data.csv')
```

```python
data.rename( columns={'Unnamed: 0':'Id'}, inplace=True )
data.head()
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm
import nltk
nltk.downloader.download('vader_lexicon')

```

```python
sia = SentimentIntensityAnalyzer()
```

```python
sia.polarity_scores('i am happy')
```

```python
result= {}
for i, row in tqdm(data.iterrows(), total=len(data)):
  text = row['review']
  myid = row['Id']
  result[myid] = sia.polarity_scores(text)
```

```python
vaders = pd.DataFrame(result).T
vaders = vaders.reset_index().rename(columns={'index':'Id'})
vaders = vaders.merge(data, how='left')
exp = vaders.head()
exp
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Load the dataset (assuming it is in a CSV file format)
#reviews_data = pd.read_csv('reviews_dataset.csv')

# Preprocess the reviews (text cleaning, tokenization, etc.)
# ...

# Analyze sentiment and assign sentiment labels to reviews
# ...

# Create an empty graph
graph = nx.Graph()

# Add nodes for reviews
for idx, review in exp.iterrows():
    review_id = review['Id']
    sentiment_label = review['compound']
    graph.add_node(review_id, label='review', sentiment=sentiment_label)

# Add nodes for sentiment labels
sentiment_labels = ['positive', 'negative']
for label in sentiment_labels:
    graph.add_node(label, label='Sentiment')

# Add edges connecting reviews to sentiment labels
for idx, review in exp.iterrows():
    review_id = review['Id']
    sentiment_label = review['compound']
    graph.add_edge(review_id, sentiment_label)

# Visualize the graph
pos = nx.spring_layout(graph)  # Positions for all nodes
node_labels = nx.get_node_attributes(graph, 'label')

nx.draw_networkx_nodes(graph, pos, node_color='lightblue')
nx.draw_networkx_labels(graph, pos, labels=node_labels)
nx.draw_networkx_edges(graph, pos)


plt.show()

plt.axis('off')
plt.show()


```
